```python
import pandas as pd

```

```python
df=pd.read_csv("traffic.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting 1: Line plot to show the trend of vehicles over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='DateTime', y='Vehicles', data=df, hue='Junction')
plt.title('Traffic Congestion Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Plotting 2: Bar plot to show the total number of vehicles at each junction
plt.figure(figsize=(10, 6))
sns.barplot(x='Junction', y='Vehicles', data=df, estimator=sum)
plt.title('Total Number of Vehicles at Each Junction')
plt.xlabel('Junction')
plt.ylabel('Total Number of Vehicles')
plt.show()


```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
# Plotting 3: Box plot to show the distribution of vehicles at each junction
plt.figure(figsize=(10, 6))
sns.boxplot(x='Junction', y='Vehicles', data=df)
plt.title('Distribution of Vehicles at Each Junction')
plt.xlabel('Junction')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Plotting 4: Correlation Heatmap to explore relationships between variables
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plotting 5: Time Series Decomposition to identify trends, seasonality, and residuals
from statsmodels.tsa.seasonal import seasonal_decompose

df.set_index('DateTime', inplace=True)
result = seasonal_decompose(df['Vehicles'], model='additive', period=1)
result.plot()
plt.suptitle('Time Series Decomposition of Traffic Congestion')
plt.show()
```

```python
# Plotting 6: Pair plot to visualize pairwise relationships between numerical variables
sns.pairplot(df, hue='Junction', height=3)
plt.suptitle('Pair Plot of Numerical Variables')
plt.show()
```

```python
# Plotting 7: Distribution plot for the 'vehicles' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Vehicles'], bins=30, kde=True)
plt.title('Distribution of Number of Vehicles')
plt.xlabel('Number of Vehicles')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting 8: Violin plot to show the distribution of vehicles at each junction
plt.figure(figsize=(12, 6))
sns.violinplot(x='Junction', y='Vehicles', data=df)
plt.title('Distribution of Vehicles at Each Junction')
plt.xlabel('Junction')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Plotting 9: Count plot to show the distribution of data points across different junctions
plt.figure(figsize=(10, 6))
sns.countplot(x='Junction', data=df)
plt.title('Count of Data Points at Each Junction')
plt.xlabel('Junction')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 9: Junction Comparison (Side-by-side Box Plots)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Junction', y='Vehicles', data=df)
plt.title('Junction Comparison by Day of the Week')
plt.xlabel('Junction')
plt.ylabel('Number of Vehicles')
plt.legend(title='Day of the Week')
plt.show()
```
